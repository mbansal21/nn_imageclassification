**Normalize the Input data**

*  we must normalize our data as it is always required in neural network models
*  we can achieve this by dividing the RGB codes with 255 (which is the maximum RGB code minus the minimum RGB code)
normalize X_train and X_test
*   make sure that the values are float so that we can get decimal points after division

```python
# normalize inputs from 0-255 to 0-1
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train /= 255
X_test /= 255
```

**Plot training images**

```python
def plot_images(X_img,y_labels,nrows,ncols,cls_pred=None):
    #plt.figure(figsize=(10,1))
    for i in range(nrows*ncols):
        plt.subplot(nrows,ncols,i+1)
        plt.imshow(X_img[i].reshape(32,32))
        # Predictions are not passed
        if cls_pred is None:
            plt.title("True: {}".format(y_labels[i]))
        # When predictions are passed, display labels + predictions
        else:
            plt.title ("True: {0}, Pred: {1}".format(y_labels[i], cls_pred[i]))
        #plt.title("Label: {}".format(y_labels[i]))
        plt.axis("off")
    plt.show() 
```

Data Splitting - Spilt Training into Train + Validation

```python
#Splitting to 20% in Val Set 
#Using random state to regenrate the whole Dataset in re-run
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train1, test_size=0.20, random_state=7)
```

**One Hot Label Encoding**

Apply One Hot Encoding to make label suitable for CNN Classification
- convert class vectors (integers) to binary class matrix
- convert y_train and y_test
- number of classes: 10
- we are doing this to use categorical_crossentropy as loss

```python
from tensorflow.keras.utils import to_categorical

print("Value before encoding:", y_train[0])

y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test1, num_classes=10)
y_val = to_categorical(y_val,num_classes=10)

print("Shape of y_train:", y_train.shape)
print("One hot encoded value of y_train:", y_train[0])
print("Shape of Validation set:", y_val.shape)
```

```python
#Let's print one image after one-hot encoding
print("Label: ", y_train[3])
plt.imshow(X_train[2], cmap='gray')
```

**Feedforward neural network**

 Where we will have inpout layer, hidden layers and one outout layer and these layer's nodes are connected in forward direction.

**Define the model**

The Sequential model is a linear stack of layers.
The model needs to know what input shape it should expect. 
For this reason, the first layer in a Sequential model (and only the first, because following layers can do automatic shape inference) needs to receive information about its input shape.

```python
from tensorflow.keras.models import Sequential
# Initialize Sequential model
model = Sequential()

# Reshape data from 2D to 1D -> 28x28 to 784
model.add(tensorflow.keras.layers.Reshape((1024,),input_shape=(32,32,)))

# Normalize the data
model.add(tensorflow.keras.layers.BatchNormalization())
#The BatchNormalization normalization layer is used to standardize inputs before or after the activation function of the previous layer.
```

First layer is added to ensure that the input layer has the right number of input features so it will specify when creating the first layer with the input_dim argument

```python

# Hidden layers
model.add(tensorflow.keras.layers.Dense(200, activation='relu', name='Layer_1'))
model.add(tensorflow.keras.layers.Dense(100, activation='relu', name='Layer_2'))

# Dropout layer
model.add(tensorflow.keras.layers.Dropout(0.5))

# Hidden layers
model.add(tensorflow.keras.layers.Dense(60, activation='relu', name='Layer_3'))
model.add(tensorflow.keras.layers.Dense(30, activation='relu', name='Layer_4'))

# Dropout layer
model.add(tensorflow.keras.layers.Dropout(0.3))
```

```python
#Output layer
model.add(tensorflow.keras.layers.Dense(10, activation='softmax', name='Output'))
```

**Compile the model**

Here we configure the model for training

We will specify an optimizer, loss function and a metrics,
You can read more about it here https://keras.io/models/sequential/

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

**Summarize the model**

model.summary() prints a summary representation of your model.

For layers with multiple outputs, multiple is displayed instead of each individual output shape due to size limitations.

```python
# Fit the model
history=model.fit(X_train, y_train,validation_data=(X_val, y_val), epochs=30, batch_size =32)
```

```python
model.summary()
```

**Final loss and accuracy**

```python
# evaluate the model
model.evaluate(X_test, y_test)
```

```python
# evaluate the model
_,train_acc = model.evaluate(X_train, y_train, verbose=0)
_,test_acc = model.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
#get b atch dynamically
def get_batch(X, y, batch_size=512):
    for i in np.arange(0, y.shape[0], batch_size):
        end = min(X.shape[0], i + batch_size)
        yield(X[i:end],y[i:end])
```

**Model without drop-out layer**

```python
#Model with 2 hidden layers
from tensorflow.keras.models import Sequential
# Initialize Sequential model
model3 = Sequential()

# Reshape data from 2D to 1D -> 28x28 to 784
model3.add(tensorflow.keras.layers.Reshape((1024,),input_shape=(32,32,)))
# Normalize the data
model3.add(tensorflow.keras.layers.BatchNormalization())


# Hidden layers
model3.add(tensorflow.keras.layers.Dense(256, activation='relu', name='Layer_1'))
model3.add(tensorflow.keras.layers.Dense(128, activation='relu', name='Layer_2'))

#Output layer
model3.add(tensorflow.keras.layers.Dense(10, activation='softmax', name='Output'))
```

```python
model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# Fit the model
history=model3.fit(X_train, y_train,validation_data=(X_val, y_val), epochs=30, batch_size =32)
```

```python
# evaluate the model
_,train_acc = model3.evaluate(X_train, y_train, verbose=0)
_,test_acc = model3.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
#so we can see that with the drop out layer we are getting better accuracy as compare to without drop layer as it is removeing the unnecessary parameters from the network
#Conclusion:
#first model is giving us the best results using batch normalization
```
